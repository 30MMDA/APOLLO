```python
import pandas as pd
import json 
from fuzzywuzzy import process
```

```python

with open("/home/imad/Desktop/brand_models.json", "r" ) as f :
    content = f.read()
```

```python

content = content.replace("'", '"')
refrece = json.loads(content)
```

```python
data = pd.read_csv("/home/imad/Desktop/Relational_Data_Base.csv")
```

```python
df = data[['description', 'name', 'brand', 'Product_type','Processor_family', 'Processor_manufacturer', 'Processor_model',
       'Processor_generation', 'Processor_cores']]
df.head(2)
```

```python
msi_list = refrece.get("MSI")
```

```python
msi_list
```

```python
refrence_df = pd.DataFrame(msi_list)

def find_best_match(name, reference_names):
    match, score = process.extractOne(name, reference_names)
    return match, score

df['reference_name'], df['match_score'] = zip(*df['name'].apply(lambda x: find_best_match(x, refrence_df['reference_name'])))

print(df)
```
